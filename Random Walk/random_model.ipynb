```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Parameters
S0 = 100        # Initial price
T = 252         # Number of days
M = 10000       # Number of paths
sigma = 1       # Standard deviation
```

```python
random_increments = np.random.normal(0, sigma, (T, M))

S = np.zeros((T, M))
S[0] = S0
```

```python
for t in range(1, T):
    S[t] = S[t-1] + random_increments[t]
```

```python
plt.plot(np.arange(T), S)
plt.title("Random Walks simulation")
plt.xlabel('Time (days)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
final_prices = S[-1, :]

plt.hist(final_prices, bins=100)
plt.title('Distribution of Final Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
