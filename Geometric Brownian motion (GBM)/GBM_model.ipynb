```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# model inputs
S0 = 100            # Inital price
mu = 0              # drift
sigma = 0.2         # Volatility
T = 1               # one year
N = 252             # path length
M = 100             # number of paths
dt = T / N          # time length
```

```python
dW = np.random.normal(0, np.sqrt(dt), (N, M))

S = np.zeros((N, M))
S[0] = S0
```

---

```python
for t in range(1, N):
    S[t] = S[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * dW[t])
```

---

```python
plt.plot(S)
plt.title("Geometric Brownian Motion simulation")
plt.xlabel('Days')
plt.ylabel('Prices')
plt.grid(True)
plt.show()
```
